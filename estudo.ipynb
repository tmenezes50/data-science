```python
import pandas as pd

tmdb = pd.read_csv("tmdb_5000_movies.csv")
tmdb.head()
tmdb.original_language.value_counts()
```

```python
tmdb.original_language.value_counts().to_frame()
```

```python
tmdb.original_language.value_counts().to_frame().reset_index()
```

```python
contagem_de_linguas = tmdb.original_language.value_counts().to_frame().reset_index()
contagem_de_linguas.columns = ["original_language", "total"]
contagem_de_linguas.head()
```

```python
import seaborn as sns
sns.barplot(x= "original_language", y = "total", data = contagem_de_linguas)
```

```python
sns.catplot(x = "original_language", kind="count", data=tmdb)
```

```python
total_por_lingua = tmdb.original_language.value_counts()
total_por_lingua.loc["en"]
```

```python
total_por_lingua = tmdb["original_language"].value_counts()
total_geral = total_por_lingua.sum()
total_de_ingles = total_por_lingua.loc['en']
total_de_resto = total_geral - total_de_ingles
print(total_de_ingles, total_de_resto)
```

```python
dados = {
    'lingua' : ['ingles', 'outros'],
    'total' : [total_de_ingles, total_de_resto]
}

pd.DataFrame(dados)
```

```python
sns.barplot(data=dados, x= 'lingua', y = 'total')
```

```python
filmes_sem_ser_ingles = tmdb.query("original_language != 'en'")
filmes_sem_ser_ingles
```

```python
outros_filmes_sem_ser_ingles = tmdb.query("original_language != 'en'").original_language.value_counts()
outros_filmes_sem_ser_ingles
```

```python
sns.catplot(x = "original_language", kind="count", data=filmes_sem_ser_ingles)
```

```python
sns.catplot(x= "original_language", kind="count", data=filmes_sem_ser_ingles, aspect= 2, palette= "GnBu_d", order=outros_filmes_sem_ser_ingles.index)
```
